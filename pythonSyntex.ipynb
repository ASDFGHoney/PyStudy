## Dictionary { }
dict는 key와 value로 구성


```python
a = {'사자' : 'lion', '호랑이':'tiger', '용':'dragon'}
print(a)
a = {'car':['bus', 'truck', 'taxi'], 'train':'ktx'}
print(a)
```

- Key

```python
key = a.keys()
print(key)
print(type(key))
print(key[0]) # 리스트 타입으로 바꿔줘야 함
```

```python
key2 = list(key)
print(type(key2))
print(key2[0])
```

- Value

```python
value = a.values()
print(value)
print(type(value))
print(value[0])# 리스트 타입으로 바꿔줘야 함
```

```python
value2 = list(value)
print(type(value2))
print(value2[0])
```

## Set (( )) or ([])
- 중복된 항목을 제거
- 순서가 없음 -> index가 없다



```python
a = set([1,2,3])
print(a)
print(type(a))

a = set((1,2,3))
print(a)
print(type(a))
```

```python
a=set((1,1,2,3,3,4))
a
```

```python
a[0] # list로 변환하면 사용 가능 ex) list(a)
```

```python
#합집합
a = set((1,2,3))
b = set([3,4,5])
c = a.union(b)
c
```

```python
#교집합
c = a.intersection(b)
c
```

```python
#차집합
c = a.difference(b)
c
```

##조건문

###if

```python
x = 6
if x >= 6:
  pass # 밑에꺼 안봄 switch랑 혼동하지 말 것
elif x == 6 and x > 3:
  print('22')
else:
  print('no')
```

##List Comprehension

```python
a = []
for i in range(1, 11, 1):
  a.append(i)
print(a)
```

```python
a = [i for i in range(1,11,1)]
a
```

```python
a = [i ** 2 for i in range(1,11,1)]
a
```

### if 활용

```python
a = [90,39,48,70,82,100]
result = [i for i in a]
a
```

```python
result = [i for i in a if i >= 80]
result
```

##함수

```python
def function1(x):
  print(x+1)
```

```python
function1(1)
function1(2)
```

```python
def function2(x):
  result = x+1
  return result
```

```python
a = function2(2)
a
```

```python
def function3(x):
  result1 = x+1
  result2 = x+2
  return result1,result2
```

```python
a, b = function3(3)
print(a)
print(b)
```

```python
def function4(*x):
  for i in x:
    print(i)

function4(1,3,5,22,3,4,'rr', 3, 2)
```
